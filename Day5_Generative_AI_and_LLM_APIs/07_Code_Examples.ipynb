## Example 1: OpenAI API — Chat Completion

````python
"""
OpenAI API — The industry standard for LLM APIs.

Key Concepts:
- client: Your connection to OpenAI's servers
- model: Which AI model to use (gpt-4o, gpt-4o-mini, etc.)
- messages: The conversation in role/content format
- temperature: Creativity control (0=focused, 1=creative)
- max_tokens: Maximum length of response
"""

from openai import OpenAI

# Step 1: Create client with API key
client = OpenAI(api_key="sk-your-openai-key-here")

# Step 2: Make a chat completion request
response = client.chat.completions.create(
    model="gpt-4o-mini",        # Model to use
    messages=[
        {
            "role": "system",     # System = Instructions for the AI
            "content": "You are a senior Python developer. Give concise, practical answers."
        },
        {
            "role": "user",       # User = The human's question
            "content": "What are the top 3 Python libraries for data science and why?"
        }
    ],
    temperature=0.7,              # Balanced creativity
    max_tokens=500                # Max response length
)

# Step 3: Extract and print the response
print("🤖 OpenAI Response:")
print(response.choices[0].message.content)

# Step 4: Check token usage (important for cost tracking!)
print(f"\n📊 Token Usage:")
print(f"  Input tokens:  {response.usage.prompt_tokens}")
print(f"  Output tokens: {response.usage.completion_tokens}")
print(f"  Total tokens:  {response.usage.total_tokens}")
print(f"  Estimated cost: ${(response.usage.prompt_tokens * 0.15 + response.usage.completion_tokens * 0.60) / 1_000_000:.6f}")
````

## Example 2: Google Gemini API

````python
"""
Google Gemini API — Google's multimodal AI model.

Key Advantages:
- Generous free tier (great for learning!)
- 2 million token context window (Gemini 1.5 Pro)
- Multimodal: Can process text, images, audio, video
- Fast and affordable
"""

import google.generativeai as genai

# Step 1: Configure with API key (get from aistudio.google.com)
genai.configure(api_key="your-gemini-api-key")

# Step 2: Create model instance
model = genai.GenerativeModel("gemini-1.5-flash")  # Fast and cheap

# Step 3: Generate content
response = model.generate_content(
    "Explain the difference between a list and a tuple in Python. "
    "Give a practical example for each."
)

# Step 4: Print response
print("🤖 Gemini Response:")
print(response.text)

# Step 5: Check token usage
print(f"\n📊 Token Usage:")
print(f"  Input tokens:  {response.usage_metadata.prompt_token_count}")
print(f"  Output tokens: {response.usage_metadata.candidates_token_count}")
````

## Example 3: Anthropic Claude API

````python
"""
Anthropic Claude API — Known for safety and long-context reasoning.

Key Advantages:
- 200K token context window
- Excellent at following complex instructions
- Strong safety guardrails
- Great for document analysis and coding
"""

import anthropic

# Step 1: Create client
client = anthropic.Anthropic(api_key="your-anthropic-key")

# Step 2: Create message
message = client.messages.create(
    model="claude-sonnet-4-20250514",  # Latest Claude model
    max_tokens=1024,
    system="You are an expert Python code reviewer. Be concise and practical.",
    messages=[
        {
            "role": "user",
            "content": "Review this code and suggest improvements:\n\n"
                       "def calc(a,b,op):\n"
                       "    if op == '+':\n"
                       "        return a+b\n"
                       "    if op == '-':\n"
                       "        return a-b\n"
                       "    if op == '*':\n"
                       "        return a*b\n"
                       "    if op == '/':\n"
                       "        return a/b"
        }
    ]
)

# Step 3: Print response
print("🤖 Claude Response:")
print(message.content[0].text)  # Note: Claude returns content as a list

# Step 4: Token usage
print(f"\n📊 Token Usage:")
print(f"  Input tokens:  {message.usage.input_tokens}")
print(f"  Output tokens: {message.usage.output_tokens}")
````

## Example 4: Groq API — Lightning Fast Inference

````python
"""
Groq API — The fastest LLM inference provider.

Key Advantages:
- 500+ tokens/second (10x faster than competitors)
- Uses custom LPU (Language Processing Unit) chips
- OpenAI-compatible API format (easy to switch!)
- Very affordable pricing
- Free tier available for learning
"""

from groq import Groq
import time

# Step 1: Create client (get key from console.groq.com)
client = Groq(api_key="gsk-your-groq-key")

# Step 2: Make request and time it
start_time = time.time()

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",  # Meta's LLaMA 3.3 70B model
    messages=[
        {
            "role": "system",
            "content": "You are a helpful coding tutor. Be concise."
        },
        {
            "role": "user",
            "content": "Write a Python function that checks if a string is a palindrome. Include error handling."
        }
    ],
    temperature=0.3,  # Low temperature for code
    max_tokens=500
)

elapsed = time.time() - start_time

# Step 3: Print response
print("🤖 Groq Response (LLaMA 3.3 70B):")
print(response.choices[0].message.content)

# Step 4: Performance metrics
tokens_generated = response.usage.completion_tokens
tokens_per_second = tokens_generated / elapsed if elapsed > 0 else 0

print(f"\n⚡ Performance:")
print(f"  Response time: {elapsed:.2f} seconds")
print(f"  Tokens generated: {tokens_generated}")
print(f"  Speed: {tokens_per_second:.0f} tokens/second")
print(f"\n📊 Token Usage:")
print(f"  Total: {response.usage.total_tokens} tokens")
````

## Example 6: Ollama — Run LLMs Locally

````python
"""
Ollama — Run LLMs locally on your machine.

Prerequisites:
1. Install Ollama: https://ollama.ai
2. Pull a model: ollama pull llama3.2
3. Start server: ollama serve

Key Advantages:
- 100% FREE (no API costs)
- 100% PRIVATE (data stays on your machine)
- No internet required
- Great for learning and prototyping
"""

import requests
import json

def chat_with_ollama(message: str, model: str = "llama3.2") -> str:
    """Send a message to locally running Ollama."""
    try:
        response = requests.post(
            "http://localhost:11434/api/chat",
            json={
                "model": model,
                "messages": [
                    {"role": "user", "content": message}
                ],
                "stream": False  # Get complete response at once
            },
            timeout=60
        )
        response.raise_for_status()
        return response.json()["message"]["content"]
    except requests.ConnectionError:
        return "❌ Error: Ollama is not running. Start it with: ollama serve"
    except Exception as e:
        return f"❌ Error: {str(e)}"

# Test the local model
print("🤖 Ollama Response (Local LLM):")
result = chat_with_ollama("What is a dictionary in Python? Give a quick example.")
print(result)
print("\n✅ This ran completely locally — no API costs, no internet needed!")
````

## Example 7: Multi-Provider Comparison

````python
"""
Compare responses from multiple providers with the same prompt.
This helps you understand the differences between models.
"""

import time
from groq import Groq
import google.generativeai as genai

# Setup clients (use the APIs you have keys for)
groq_client = Groq(api_key="your-groq-key")
genai.configure(api_key="your-gemini-key")

PROMPT = "Explain what an API is to a 10-year-old. Use an analogy."

def query_groq(prompt: str) -> dict:
    """Query Groq API."""
    start = time.time()
    response = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=200
    )
    elapsed = time.time() - start
    return {
        "provider": "Groq (LLaMA 70B)",
        "response": response.choices[0].message.content,
        "time": elapsed,
        "tokens": response.usage.total_tokens
    }

def query_gemini(prompt: str) -> dict:
    """Query Gemini API."""
    start = time.time()
    model = genai.GenerativeModel("gemini-1.5-flash")
    response = model.generate_content(prompt)
    elapsed = time.time() - start
    return {
        "provider": "Gemini 1.5 Flash",
        "response": response.text,
        "time": elapsed,
        "tokens": response.usage_metadata.total_token_count
    }

# Run comparison
print(f"📝 Prompt: '{PROMPT}'")
print("=" * 70)

results = [query_groq(PROMPT), query_gemini(PROMPT)]

for result in results:
    print(f"\n🤖 {result['provider']}")
    print(f"⏱️ Time: {result['time']:.2f}s | 📊 Tokens: {result['tokens']}")
    print(f"💬 Response:\n{result['response'][:300]}")
    print("-" * 70)

# Summary table
print("\n📊 COMPARISON SUMMARY")
print(f"{'Provider':<25} {'Time':<10} {'Tokens':<10}")
print("-" * 45)
for r in results:
    print(f"{r['provider']:<25} {r['time']:.2f}s     {r['tokens']}")
````

## Example 8: Token Counter & Cost Calculator

````python
"""
Production Token Counter & Cost Calculator
Use this to estimate costs before building your AI app.
"""

import tiktoken

class TokenCostCalculator:
    """Calculate token counts and costs for different LLM providers."""
    
    # Pricing per 1M tokens (as of 2024)
    PRICING = {
        "gpt-4o": {"input": 2.50, "output": 10.00},
        "gpt-4o-mini": {"input": 0.15, "output": 0.60},
        "claude-3.5-sonnet": {"input": 3.00, "output": 15.00},
        "gemini-1.5-flash": {"input": 0.075, "output": 0.30},
        "gemini-1.5-pro": {"input": 1.25, "output": 5.00},
        "groq-llama-70b": {"input": 0.59, "output": 0.79},
    }
    
    def __init__(self):
        self.encoder = tiktoken.encoding_for_model("gpt-4o")
    
    def count_tokens(self, text: str) -> int:
        """Count tokens in a text string."""
        return len(self.encoder.encode(text))
    
    def calculate_cost(self, input_tokens: int, output_tokens: int, model: str) -> float:
        """Calculate cost for a single API call."""
        if model not in self.PRICING:
            raise ValueError(f"Unknown model: {model}")
        
        pricing = self.PRICING[model]
        input_cost = (input_tokens / 1_000_000) * pricing["input"]
        output_cost = (output_tokens / 1_000_000) * pricing["output"]
        return input_cost + output_cost
    
    def monthly_projection(self, input_tokens: int, output_tokens: int,
                           daily_requests: int, model: str) -> dict:
        """Project monthly costs."""
        cost_per_request = self.calculate_cost(input_tokens, output_tokens, model)
        daily_cost = cost_per_request * daily_requests
        monthly_cost = daily_cost * 30
        
        return {
            "model": model,
            "cost_per_request": cost_per_request,
            "daily_requests": daily_requests,
            "daily_cost": daily_cost,
            "monthly_cost": monthly_cost
        }

# Usage example
calc = TokenCostCalculator()

# Count tokens in sample text
sample = "Explain machine learning to me like I'm a beginner."
tokens = calc.count_tokens(sample)
print(f"📝 '{sample}'")
print(f"📊 Token count: {tokens}")

# Monthly cost projection
print("\n💰 Monthly Cost Projection (10,000 requests/day)")
print("=" * 60)
print(f"{'Model':<25} {'Per Request':<15} {'Daily':<12} {'Monthly':<12}")
print("-" * 60)

for model in calc.PRICING.keys():
    proj = calc.monthly_projection(
        input_tokens=500,
        output_tokens=300,
        daily_requests=10000,
        model=model
    )
    print(f"{model:<25} ${proj['cost_per_request']:.6f}    "
          f"${proj['daily_cost']:.2f}     ${proj['monthly_cost']:.2f}")
````
