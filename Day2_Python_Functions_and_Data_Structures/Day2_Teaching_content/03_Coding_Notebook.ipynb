## 1. Functions & Variables Scope
Functions let you bundle code for reuse. Let's write a reusable function to calculate food taxes (GST) for different restaurants.

```python
# Define the function
def calculate_total_amount(base_price: float, discount: float = 0.0, gst_rate: float = 0.05) -> float:
    """
    Calculates the final payable amount applying a discount first, then adding GST.
    """
    discounted_price = base_price - discount
    gst_amount = discounted_price * gst_rate
    final_total = discounted_price + gst_amount
    return final_total

# Call the function with keyword arguments
bill_1 = calculate_total_amount(base_price=250.0, discount=30.0)
print(f'Bill 1 (5% GST): ₹{bill_1:.2f}')

# Call the function with custom GST (e.g., 18% for AC luxury dining)
bill_2 = calculate_total_amount(base_price=500.0, discount=50.0, gst_rate=0.18)
print(f'Bill 2 (18% GST): ₹{bill_2:.2f}')
```

## 2. Python Lists (Ordered & Mutable)
Lists are used to store multiple items in a single variable. They are indexable (starting from 0) and mutable (can be changed).

```python
# Create an initial cart list
cart = ['Biryani', 'Samosa', 'Cold Drink']
print(f'Initial Cart: {cart}')

# 1. Adding items
cart.append('Ice Cream') # Adds to the end
cart.insert(1, 'Garlic Bread') # Inserts at index 1
print(f'After Additions: {cart}')

# 2. Accessing items
print(f'First Item: {cart[0]}')
print(f'Last Item: {cart[-1]}')

# 3. Modifying items
cart[2] = 'Special Samosa' # Mutating the list value
print(f'After Mutation: {cart}')

# 4. Removing items
removed_item = cart.pop(3) # Removes and returns item at index 3
print(f'Popped Item: {removed_item}')
cart.remove('Special Samosa') # Removes first occurrence of this value
print(f'Final Cart: {cart}')
```

## 3. Python Tuples (Ordered & Immutable)
Tuples are identical to lists, except they are **immutable**—once defined, they cannot be edited. They protect critical constants (like coordinate configs).

```python
# Define restaurant coordinate tuple
nagpur_depot_coords = (21.1458, 79.0882)
print(f'GPS Coordinates: {nagpur_depot_coords}')

# Accessing values
print(f'Latitude: {nagpur_depot_coords[0]}')

# Tuple Unpacking
lat, lng = nagpur_depot_coords
print(f'Unpacked -> Lat: {lat}, Lng: {lng}')

# Attempting to change a tuple throws an error:
try:
    nagpur_depot_coords[0] = 22.0
except TypeError as e:
    print(f'❌ Error Caught: {e}')
```

## 4. Python Sets (Unordered & Unique)
Sets store unique, unordered elements. They provide $O(1)$ fast lookup and support set operations like Intersection, Union, and Difference.

```python
# Set creation (duplicates are automatically filtered)
customer_ids = {1001, 1002, 1003, 1001, 1004}
print(f'Unique Customers set: {customer_ids}')

# Membership Checking (O(1) complexity)
print(f'Is 1001 in system? {1001 in customer_ids}')

# Set Algebra
morning_delivery_agents = {'Ramesh', 'Suresh', 'Ganesh'}
evening_delivery_agents = {'Suresh', 'Ganesh', 'Dinesh'}

# 1. Intersection (Agents working both shifts)
both_shifts = morning_delivery_agents & evening_delivery_agents
print(f'Both Shifts (Intersection): {both_shifts}')

# 2. Union (All unique agents active today)
all_agents = morning_delivery_agents | evening_delivery_agents
print(f'All Active Agents (Union): {all_agents}')

# 3. Difference (Morning agents who don't work evening shift)
morning_only = morning_delivery_agents - evening_delivery_agents
print(f'Morning Only (Difference): {morning_only}')
```

## 5. Python Dictionaries (Key-Value Pairs)
Dictionaries map keys to values. They offer fast lookups and are used to represent structural entities (like restaurant menus or user accounts).

```python
# Create menu dictionary
menu = {
    'Samosa': 15,
    'Tea': 10,
    'Poha': 20
}
print(f'Menu: {menu}')

# 1. Relational Lookup
print(f'Price of Poha: ₹{menu["Poha"]}')

# 2. Safe lookups using get() to avoid KeyErrors
price_of_chips = menu.get('Chips', 0) # Fallbacks to 0 if key not found
print(f'Price of Chips (Safe Lookup): ₹{price_of_chips}')

# 3. Add / Update Key-Value pairs
menu['Biscuit'] = 5   # Adds new key
menu['Samosa'] = 18   # Modifies existing key price
print(f'Updated Menu: {menu}')

# 4. Traversing items
print('\n--- Canteen Menu Price Sheet ---')
for item, price in menu.items():
    print(f'- {item}: ₹{price}')
```

## 6. List Comprehensions
List comprehensions provide a concise way to create lists from existing sequences without writing extensive, multi-line `for` loops.

```python
prices = [15, 10, 150, 120, 50, 5]

# Task: Create a new list containing items that cost more than 40 INR

# Method A: Traditional Loop
expensive_prices_a = []
for p in prices:
    if p > 40:
        expensive_prices_a.append(p)
print(f'Traditional Loop result: {expensive_prices_a}')

# Method B: List Comprehension (Concise, Pythonic, and Faster!)
expensive_prices_b = [p for p in prices if p > 40]
print(f'List Comprehension result: {expensive_prices_b}')
```

## 🛠️ Mini Project: Student Marksheet and Analytics System

### Scenario
A school teacher in Nagpur needs to analyze student exam marks. The raw data contains a list of student records (each represented as a dictionary). We need to write a utility to:
1. Calculate the average score of the class.
2. Identify the top-performing student.
3. List students who passed the exam (score >= 40) using list comprehensions.

```python
# Raw Student Records
STUDENTS = [
    {'name': 'Amit', 'roll_no': 101, 'score': 85},
    {'name': 'Sneha', 'roll_no': 102, 'score': 38},
    {'name': 'Rahul', 'roll_no': 103, 'score': 92},
    {'name': 'Priya', 'roll_no': 104, 'score': 45},
    {'name': 'Ganesh', 'roll_no': 105, 'score': 32}
]

def analyze_student_marks(student_list):
    # 1. Calculate Average
    total_score = sum([s['score'] for s in student_list])
    class_average = total_score / len(student_list)
    
    # 2. Identify Top Scorer
    top_student = max(student_list, key=lambda s: s['score'])
    
    # 3. List of Students who passed (Score >= 40) using List Comprehension
    passed_students = [s['name'] for s in student_list if s['score'] >= 40]
    
    return class_average, top_student, passed_students

# Run Analytics
avg, top, passed = analyze_student_marks(STUDENTS)

print('=== STUDENT CLASS ANALYTICS ===')
print(f'Class Average Score : {avg:.2f}')
print(f'Class Top Performer : {top["name"]} (Roll No: {top["roll_no"]}) with {top["score"]}/100')
print(f'Passed Students     : {", ".join(passed)}')
print('================================')
```

## 🚀 Industry Project: Zomato Cart Billing & Dynamic Discount Engine

### Scenario
Zomato needs a robust billing engine for its checkout page. The billing system must satisfy the following constraints:
1. **Menu Mapping:** Read item names from a customer cart and fetch their prices from a restaurant menu catalog.
2. **Quantities Aggregation:** Since customers can add the same item multiple times in their cart list, aggregate these items to calculate unique quantities.
3. **OutOfStock Protection:** If an item in the cart is missing from the menu, flag it as "unresolved" and do not calculate it in the billing amount.
4. **Promo Codes Processing:** Validate promo codes dynamically. Active promo codes are managed via sets.
5. **Clean Invoice Printing:** Print a clean structure with item breakdowns, subtotal, discount, GST, and net payable amount.

```python
# 1. Menu Catalog (Dictionary)
ZOMATO_MENU = {
    'veg_burger': 90,
    'chicken_biryani': 220,
    'garlic_bread': 80,
    'cold_coffee': 60,
    'fries': 50
}

# 2. Active Promo Codes (Set)
ACTIVE_PROMO_CODES = {'ZOMATO50', 'BIRYANI_FEAST', 'FREE_SHIPPING'}

def zomato_billing_engine(cart: list, promo_code: str = None) -> dict:
    """
    Accepts a cart list and promo code. Processes items, handles quantity counts,
    excludes out of stock items, applies promo code logic, adds GST, and outputs invoice details.
    """
    # Aggregate item frequencies (Quantity mapping)
    cart_quantities = {}
    unresolved_items = []
    
    for raw_item in cart:
        item = raw_item.lower().strip().replace(' ', '_')
        if item in ZOMATO_MENU:
            cart_quantities[item] = cart_quantities.get(item, 0) + 1
        else:
            unresolved_items.append(raw_item)

    # Calculate Subtotal
    subtotal = 0.0
    item_breakdown = []
    for item, qty in cart_quantities.items():
        price_per_unit = ZOMATO_MENU[item]
        item_total = price_per_unit * qty
        subtotal += item_total
        item_breakdown.append({
            'item_name': item.replace('_', ' ').title(),
            'qty': qty,
            'unit_price': price_per_unit,
            'total_price': item_total
        })
        
    # Apply Promo Code Discount (Dynamic conditions)
    discount_amount = 0.0
    promo_applied = False
    
    if promo_code and promo_code.upper().strip() in ACTIVE_PROMO_CODES:
        code = promo_code.upper().strip()
        if code == 'ZOMATO50':
            discount_amount = subtotal * 0.50  # 50% Flat discount
            # Limit discount to maximum ₹100 in line with standard coupons
            if discount_amount > 100:
                discount_amount = 100.0
            promo_applied = True
        elif code == 'BIRYANI_FEAST' and 'chicken_biryani' in cart_quantities:
            # 30% off on the entire cart if Chicken Biryani is ordered
            discount_amount = subtotal * 0.30
            promo_applied = True
        elif code == 'FREE_SHIPPING':
            # Standard shipping fee (simulated here as ₹30 coupon credit)
            discount_amount = 30.0
            promo_applied = True
            
    # Compute taxes (5% GST)
    taxable_value = subtotal - discount_amount
    gst_amount = taxable_value * 0.05
    final_payable = taxable_value + gst_amount
    
    # Build Result Invoice structure
    return {
        'items_breakdown': item_breakdown,
        'subtotal': subtotal,
        'discount': discount_amount,
        'promo_applied': promo_code if promo_applied else None,
        'gst': gst_amount,
        'payable': final_payable,
        'unresolved_items': list(set(unresolved_items))  # Return unique missing items
    }

# Simulate customer ordering
user_cart = [
    'chicken_biryani',
    'veg_burger',
    'chicken_biryani',
    'cold_coffee',
    'paneer_tikka_masala'  # Out of stock item
]
coupon = 'ZOMATO50'

# Run Billing Engine
invoice = zomato_billing_engine(user_cart, promo_code=coupon)

# Print output invoice structure
print('============== ZOMATO INVOICE ==============')
print(f'Promo Code Submitted: {coupon}')
if invoice['promo_applied']:
    print(f'Coupon Status: SUCCESS ({invoice["promo_applied"]} applied!)')
else:
    print('Coupon Status: FAILED or Invalid Coupon')
print('-------------------------------------------')
for item in invoice['items_breakdown']:
    print(f'{item["item_name"]:18} x {item["qty"]} @ ₹{item["unit_price"]:3} = ₹{item["total_price"]:3}')
print('-------------------------------------------')
print(f'Subtotal             : ₹{invoice["subtotal"]:.2f}')
print(f'Discount             : -₹{invoice["discount"]:.2f}')
print(f'GST (5%)             : ₹{invoice["gst"]:.2f}')
print(f'Net Amount Payable   : ₹{invoice["payable"]:.2f}')
if invoice['unresolved_items']:
    print('-------------------------------------------')
    print(f'⚠️ Items out-of-stock (Excluded): {", ".join(invoice["unresolved_items"])}')
print('============================================')
```
