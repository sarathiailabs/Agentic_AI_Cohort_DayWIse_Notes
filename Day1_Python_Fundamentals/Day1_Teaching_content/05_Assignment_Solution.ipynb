## ➕ 1. Project 1 Solution: Interactive Calculator
Features:
- Formatted float conversions with `try-except` to prevent crashes when strings are typed.
- Divisor zero verification.
- Two decimal places print styling.

```python
def run_calculator():
    print("\n--- 📱 INTERACTIVE CALCULATOR ---")
    try:
        # Validate numeric inputs using try-except
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
    except ValueError:
        print("❌ Error: Please input valid numbers (integers or decimals).")
        return
    
    operator = input("Enter operator (+, -, *, /): ").strip()
    
    # Logic and execution block
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        # Check for division by zero
        if num2 == 0:
            print("❌ Error: Division by zero is mathematically undefined.")
            return
        result = num1 / num2
    else:
        print(f"❌ Error: '{operator}' is not a supported operator.")
        return
        
    # Output formatted to 2 decimal places
    print(f"🎉 Result: {num1} {operator} {num2} = {result:.2f}")

# Test execution (uncomment to run)
# run_calculator()
```

## 🎯 2. Project 2 Solution: Smart Number Guessing Game
Features:
- Random number generation via `random.randint`.
- Attempts counter tracker.
- Replay loop option after winning.

```python
import random

def run_guessing_game():
    while True:
        print("\n--- 🎯 NUMBER GUESSING GAME ---")
        print("I have chosen a number between 1 and 100. Can you guess it?")
        
        target_number = random.randint(1, 100)
        attempts = 0
        
        # Game loop
        while True:
            try:
                guess = int(input("Enter your guess: "))
            except ValueError:
                print("❌ Invalid input! Please enter a valid integer.")
                continue
            
            attempts += 1
            
            if guess > target_number:
                print("📉 Too High! Try again.")
            elif guess < target_number:
                print("📈 Too Low! Try again.")
            else:
                print(f"🎉 CONGRATULATIONS! You guessed it in {attempts} attempts!")
                break
        
        # Replay prompt
        play_again = input("Do you want to play again? (y/n): ").strip().lower()
        if play_again != 'y':
            print("Thanks for playing! 👋")
            break

# Test execution (uncomment to run)
# run_guessing_game()
```

## 🎓 3. Project 3 Solution: Student Grade Analyzer
Features:
- Subject-wise threshold verification (>= 35 marks filter).
- Interactive average compilation.
- Grade division mapping.
- Clear CLI table formatting layout.

```python
def run_grade_analyzer():
    print("\n--- 📝 STUDENT GRADE ANALYZER ---")
    student_name = input("Enter student name: ").strip()
    
    # Input collection & range checks
    try:
        math_marks = float(input("Enter Mathematics marks (0-100): "))
        science_marks = float(input("Enter Science marks (0-100): "))
        english_marks = float(input("Enter English marks (0-100): "))
        
        # Validation of ranges
        if not (0 <= math_marks <= 100 and 0 <= science_marks <= 100 and 0 <= english_marks <= 100):
            print("❌ Error: Marks must be between 0 and 100.")
            return
    except ValueError:
        print("❌ Error: Please enter valid numeric marks.")
        return
    
    # Pass status logic: Must get >= 35 in each subject
    passed_all = (math_marks >= 35) and (science_marks >= 35) and (english_marks >= 35)
    
    # Average Calculation
    total_marks = math_marks + science_marks + english_marks
    average_score = total_marks / 3
    
    # Grade assignment logic
    if not passed_all:
        division = "Fail (Failed one or more subjects)"
        status = "FAILED"
    else:
        status = "PASSED"
        if average_score >= 60:
            division = "First Division"
        elif average_score >= 50:
            division = "Second Division"
        elif average_score >= 35:
            division = "Third Division"
        else:
            division = "Fail"
            status = "FAILED"
            
    # Report Card layout printout
    print("\n========================================")
    print(f"            REPORT CARD: {student_name.upper()}")
    print("========================================")
    print(f"Mathematics : {math_marks}/100   (Pass status: {'Pass' if math_marks >= 35 else 'Fail'})")
    print(f"Science     : {science_marks}/100   (Pass status: {'Pass' if science_marks >= 35 else 'Fail'})")
    print(f"English     : {english_marks}/100   (Pass status: {'Pass' if english_marks >= 35 else 'Fail'})")
    print("----------------------------------------")
    print(f"Average Score : {average_score:.2f}%")
    print(f"Status        : {status}")
    print(f"Grade Division: {division}")
    print("========================================")

# Test execution (uncomment to run)
# run_grade_analyzer()
```
