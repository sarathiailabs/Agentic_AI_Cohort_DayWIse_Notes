## 💾 2. Variables & Data Types

Variables are labels pointing to values in computer memory. Python handles type assignment automatically (Dynamic Typing).

```python
# Defining variables of different types
customer_name = "Rakesh Kumar"  # String (str)
samosa_quantity = 5             # Integer (int)
samosa_price = 15.50            # Float (float)
bill_paid = False               # Boolean (bool)

# Checking their types
print(f"customer_name type: {type(customer_name)}")
print(f"samosa_quantity type: {type(samosa_quantity)}")
print(f"samosa_price type: {type(samosa_price)}")
print(f"bill_paid type: {type(bill_paid)}")
```

--- 
## ➕ 3. Operators & Expressions

Operators allow us to perform mathematical operations, compare values, and check combined conditions.

```python
# Arithmetic Operators
subtotal = samosa_quantity * samosa_price
discount = 10.0
final_total = subtotal - discount

print(f"Subtotal: ₹{subtotal}")
print(f"Final Total (after discount): ₹{final_total}")

# Floor Division and Modulo
split_count = 3
per_person_share = final_total // split_count  # Integer division
leftover_change = final_total % split_count    # Remainder
print(f"Per person share: ₹{per_person_share}")
print(f"Leftover change: ₹{leftover_change}")
```

## 🚦 4. Conditions (`if-elif-else`)

Conditions steer our program based on logical tests. In Python, blocks are defined by **indentation** (4 spaces).

```python
attendance_percentage = 78
has_paid_fees = True

# Check exam eligibility
if attendance_percentage >= 75 and has_paid_fees:
    print('✅ Student is eligible to receive Hall Ticket.')
elif attendance_percentage < 75 and has_paid_fees:
    print('❌ Ineligible: Low attendance (required >= 75%).')
else:
    print('❌ Ineligible: Fees payment pending.')
```

## 🔄 5. Loops (`for` and `while`)

Loops execute a code block multiple times.

```python
print("=== For Loop (Iteration over a range) ===")
for count in range(1, 4):
    print(f"Serving Samosa plate number {count}")

print("\n=== While Loop (Iterating while condition is met) ===")
queue_size = 3
while queue_size > 0:
    print(f"Processing student in queue. Remaining queue: {queue_size}")
    queue_size -= 1  # Crucial! Decrement variables to prevent infinite loop
```

## 🛠️ 6. Mini-Project: Interactive CLI Calculator

Let's write a simple calculator that reads user entries from the console, checks the operation, calculates the result, and prints it.

```python
# Code template for CLI Calculator
def run_calculator():
    print("=== Simple CLI Calculator ===")
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        operator = input("Enter operator (+, -, *, /): ").strip()

        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("❌ Error: Division by zero is not allowed.")
                return
            result = num1 / num2
        else:
            print("❌ Error: Invalid operator entered.")
            return

        print(f"🎉 Result: {num1} {operator} {num2} = {result}")
    except ValueError:
        print("❌ Error: Please enter valid numbers.")

# Run the calculator (uncomment to test locally)
# run_calculator()
```

## 🏭 7. Industry Challenge: Canteen Bill Generator

**Challenge**: Ramesh Kaka's canteen needs an billing generator that: 
- Prompts for the customer's name.
- Asks how many items they ordered (Samosas @ ₹15, Tea @ ₹10, Maggi @ ₹30).
- Calculates Subtotal.
- Applies a 10% discount if subtotal exceeds ₹200.
- Generates a beautifully formatted terminal invoice.

```python
# Write your Canteen Bill Generator logic here!
def generate_canteen_bill():
    # 1. Ask for inputs
    customer = input("Enter customer name: ")
    samosas = int(input("Enter samosas quantity (₹15): ") or 0)
    teas = int(input("Enter tea quantity (₹10): ") or 0)
    maggis = int(input("Enter maggi quantity (₹30): ") or 0)

    # 2. Compute prices
    samosa_total = samosas * 15
    tea_total = teas * 10
    maggi_total = maggis * 30
    subtotal = samosa_total + tea_total + maggi_total

    # 3. Dynamic Discount condition
    discount = 0.0
    if subtotal > 200:
        discount = subtotal * 0.10
    
    final_bill = subtotal - discount

    # 4. Print clean receipt
    print("\n====================================")
    print(f"🧾 INVOICE - RAMESH KAKA'S CANTEEN")
    print("====================================")
    print(f"Customer: {customer}")
    print("------------------------------------")
    if samosas > 0: print(f"Samosas   x {samosas:<3} : ₹{samosa_total}")
    if teas > 0:    print(f"Tea       x {teas:<3} : ₹{tea_total}")
    if maggis > 0:  print(f"Maggi     x {maggis:<3} : ₹{maggi_total}")
    print("------------------------------------")
    print(f"Subtotal         : ₹{subtotal}")
    print(f"Discount (10%)   : -₹{discount:.2f}")
    print(f"Total Bill       : ₹{final_bill:.2f}")
    print("====================================")

# Run the generator (uncomment to test locally)
# generate_canteen_bill()
```

## 💡 8. Summary & Key Takeaways
- **Variables** hold temporary data; use meaningful snake_case names (e.g. `samosa_price`).
- Always cast your `input()` variables (e.g. `int()`, `float()`) before attempting mathematical operations.
- Be extremely careful with **indentations**. 4 spaces are the standard in Python; mixing spaces and tabs causes `IndentationError`.
- Always include an update operation (like `count -= 1`) inside a `while` loop to prevent freezing your terminal in an infinite cycle.
