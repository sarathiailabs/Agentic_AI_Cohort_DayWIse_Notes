## 🛠️ Setup and Authentication

First, let's load our API keys from the `.env` file.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# Load .env file
load_dotenv()

# Initialize OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
print("OpenAI Client initialized successfully!")
```

--- 
## 1. Zero-Shot vs Few-Shot Prompting

Let's see how the model behaves when asked to classify text without examples (Zero-Shot) versus with examples (Few-Shot).

```python
# Zero-Shot Example: Custom sentiment analysis tagging
zero_shot_prompt = """
Classify the sentiment of the following customer review into one of these tags: [DELIVERY-DELAY], [PRODUCT-DAMAGE], [SUPPORT-EXCELLENT].

Review: "The delivery driver arrived 3 hours after the expected time, but the product was undamaged and support refunded my fee."
Tag:
"""

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": zero_shot_prompt}],
    temperature=0
)

print("--- Zero-Shot Output ---")
print(response.choices[0].message.content.strip())
```

```python
# Few-Shot Example: Direct patterns guiding output
few_shot_prompt = """
Classify the sentiment of the customer reviews into one of these tags: [DELIVERY-DELAY], [PRODUCT-DAMAGE], [SUPPORT-EXCELLENT].

Review: "My package was left in the rain and the box was soaked."
Tag: [DELIVERY-DELAY]

Review: "The customer service rep solved my issue in 2 minutes. Excellent!"
Tag: [SUPPORT-EXCELLENT]

Review: "The glass was shattered when I opened the box."
Tag: [PRODUCT-DAMAGE]

Review: "The delivery driver arrived 3 hours after the expected time, but the product was undamaged and support refunded my fee."
Tag:
"""

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": few_shot_prompt}],
    temperature=0
)

print("--- Few-Shot Output ---")
print(response.choices[0].message.content.strip())
```

## 2. Role Prompting

By giving the LLM a persona, we constrain the scope of its responses.

```python
topic = "Docker containerization"

generic_prompt = f"Explain {topic} in brief."

role_prompt = f"""
You are an elite, cynical systems engineer who has been running Linux servers for 20 years.
Explain {topic} in brief, maintaining your persona.
"""

response_generic = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": generic_prompt}],
    temperature=0.7
)

response_role = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": role_prompt}],
    temperature=0.7
)

print("=== Generic Output ===")
print(response_generic.choices[0].message.content.strip())
print("\n=== Role Prompted Output ===")
print(response_role.choices[0].message.content.strip())
```

## 3. Chain of Thought (CoT) Prompting

Let's see how reasoning steps improve performance on logic puzzles.

```python
puzzle = "If a room contains 10 people, and each person shakes hands with every other person exactly once, how many handshakes take place?"

standard_prompt = f"{puzzle} Give only the final answer."
cot_prompt = f"{puzzle} Think step-by-step and write out your mathematical reasoning steps before outputting the final number."

response_std = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": standard_prompt}],
    temperature=0
)

response_cot = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": cot_prompt}],
    temperature=0
)

print("=== Standard Output ===")
print(response_std.choices[0].message.content.strip())
print("\n=== CoT Output ===")
print(response_cot.choices[0].message.content.strip())
```
