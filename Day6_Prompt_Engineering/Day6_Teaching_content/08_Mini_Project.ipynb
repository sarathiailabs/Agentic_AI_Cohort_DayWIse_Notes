# 📓 Day 6 Mini-Project: LinkedIn Post Generator

In this mini-project, we will build a command-line utility and Jupyter workflow to convert technical ideas or project highlights into highly engaging, formatted LinkedIn posts using **Few-Shot Prompting** and **Role Prompting**.

---

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

### 1. Defining the Copywriter Role & Templates

We define a custom system instruction that forces the LLM to structure its posts with a Hook, Bulleted lists, and a Clear CTA.

```python
system_role = """
You are a professional LinkedIn content strategist and tech copywriter.
Your writing is direct, technical, yet highly engaging. You avoid corporate clichés like 'thrilled to announce' or 'revolutionizing the industry'.

Follow this exact layout for every post:
1. Hook: An attention-grabbing first line (starts with a relevant emoji).
2. Problem/Context: Explain why this topic matters.
3. Bullet Points: 3 clear, punchy takeaways or tips.
4. Pro-tip: A brief, highly tactical piece of advice.
5. CTA: An open-ended question that encourages engagement.
"""

few_shot_examples = """
Example 1:
Topic: Why you should stop using global Python installations.
Post:
📦 Stop installing Python packages globally!

Every time you run 'pip install' outside a virtual environment, you are risking system instability.

Here is why you need a virtual environment today:
- Prevents version conflicts between different projects
- Keeps your host operating system clean and secure
- Ensures your teammate can replicate your code using a single requirements.txt file

💡 Pro-tip: Use 'uv venv' to create environments in less than 0.1 seconds.

What is your preferred environment manager? Let me know below! 👇
"""
```

### 2. Generating the Post

```python
def generate_linkedin_post(topic: str) -> str:
    user_prompt = f"""
    {few_shot_examples}
    
    Topic: {topic}
    Post:
    """
    
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": system_role},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.7
    )
    return response.choices[0].message.content.strip()

topic_input = "Learning Git rebase vs merge"
post_output = generate_linkedin_post(topic_input)
print("=== Generated LinkedIn Post ===")
print(post_output)
```
