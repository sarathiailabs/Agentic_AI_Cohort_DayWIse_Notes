## Step 3: Implement Feature Branching & Induce Conflict
We will simulate parallel development. We will create a `feature-billing` branch to add discount features. Meanwhile, we will make a conflicting edit on the `main` branch.

```python
# Create feature-billing branch
!git checkout -b feature-billing
```

```python
# Modify kiosk.py to include tax and flat discounts in feature branch
feature_code = """# Coimbatore Retail Billing Kiosk
import os

def run_kiosk():
    print("=== COIMBATORE SMART RETAIL KIOSK ===")
    price = float(input("Enter base item price (INR): "))
    quantity = int(input("Enter quantity: "))
    total = price * quantity
    
    # Feature-billing: Flat ₹20 discount for items total > ₹100
    discount = 20.0 if total > 100 else 0.0
    net_total = total - discount
    
    print(f"Gross Total: INR {total}")
    print(f"Discount Applied: INR {discount}")
    print(f"Net Payable: INR {net_total}")

if __name__ == '__main__':
    run_kiosk()
"""

with open("kiosk.py", "w") as f:
    f.write(feature_code)

print("Modified kiosk.py on feature-billing branch")
```

```python
# Stage and commit changes on feature-billing
!git add kiosk.py
!git commit -m "feat: integrate flat rate retail discounts logic"
```

```python
# Switch back to main branch
!git checkout main
```

```python
# Edit kiosk.py on main branch with conflicting discount strategy (e.g. 5% tax and 10% discount)
main_conflicting_code = """# Coimbatore Retail Billing Kiosk
import os

def run_kiosk():
    print("=== COIMBATORE SMART RETAIL KIOSK ===")
    price = float(input("Enter base item price (INR): "))
    quantity = int(input("Enter quantity: "))
    total = price * quantity
    
    # Main Branch: 10% promotional discount
    discount = total * 0.10
    net_total = total - discount
    
    print(f"Gross Total: INR {total}")
    print(f"Promotional Discount (10%): INR {discount}")
    print(f"Net Payable: INR {net_total}")

if __name__ == '__main__':
    run_kiosk()
"""

with open("kiosk.py", "w") as f:
    f.write(main_conflicting_code)

print("Modified kiosk.py on main branch to create conflict")
```

```python
# Commit conflicting main branch changes
!git add kiosk.py
!git commit -m "feat: implement 10% promotional discount scheme"
```

## Step 4: Merge & Resolve Conflict
Let's run the merge command to trigger the conflict, verify it is blocked, and then resolve the code.

```python
# Attempt merge
!git merge feature-billing
```

```python
# Check conflict markers
with open("kiosk.py", "r") as f:
    print(f.read())
```

```python
# Write the resolved code, choosing the 10% promotional discount
resolved_code = """# Coimbatore Retail Billing Kiosk
import os

def run_kiosk():
    print("=== COIMBATORE SMART RETAIL KIOSK ===")
    price = float(input("Enter base item price (INR): "))
    quantity = int(input("Enter quantity: "))
    total = price * quantity
    
    # Resolved: 10% promotional discount preferred by marketing
    discount = total * 0.10
    net_total = total - discount
    
    print(f"Gross Total: INR {total}")
    print(f"Discount (10%): INR {discount}")
    print(f"Net Payable: INR {net_total}")

if __name__ == '__main__':
    run_kiosk()
"""

with open("kiosk.py", "w") as f:
    f.write(resolved_code)

print("Conflict resolved manually in kiosk.py")
```

```python
# Complete merge
!git add kiosk.py
!git commit -m "merge: resolve billing conflicts by selecting 10% promo rate"
```
