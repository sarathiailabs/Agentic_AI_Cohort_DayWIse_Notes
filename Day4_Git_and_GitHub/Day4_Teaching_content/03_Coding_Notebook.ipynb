## Part 2: File Tracking & Committing
Let's create a core application file, check its tracking status, stage it, and commit it.

```python
# Create a basic Python script
with open("app.py", "w") as f:
    f.write('print("=== COIMBATORE CANTEEN BILLING v1.0.0 ===")\n')
    f.write('price = 15\n')
    f.write('count = 3\n')
    f.write('print(f"Total: {price * count}")\n')

print("Created app.py")
```

```python
# Check untracked files status
!git status
```

```python
# Stage the file
!git add app.py

# Verify the file is staged
!git status
```

```python
# Commit the staged snapshot
!git commit -m "feat: create core canteen billing structure"
```

```python
# Check history log
!git log --oneline
```

## Part 3: Branching & Merge Conflicts
We will now create a parallel branch to develop a discount feature. We will modify the same lines of code in both `main` and `feature-discount` branches to trigger a merge conflict.

```python
# Create and checkout new branch
!git checkout -b feature-discount

# List current branches
!git branch
```

```python
# Append discount function to app.py on feature-discount branch
with open("app.py", "a") as f:
    f.write('def apply_discount(total): return total - 10  # ₹10 flat discount\n')
    f.write('print(f"Final with discount: {apply_discount(price * count)}")\n')

# Stage and commit
!git add app.py
!git commit -m "feat: add flat discount logic"
```

```python
# Switch back to main branch
!git checkout main
```

```python
# Append a DIFFERENT discount logic to app.py on the main branch
with open("app.py", "a") as f:
    f.write('def apply_discount(total): return total * 0.90  # 10% percentage discount\n')
    f.write('print(f"Final with percentage discount: {apply_discount(price * count)}")\n')

# Stage and commit on main
!git add app.py
!git commit -m "feat: add percentage discount logic"
```

### 3.1: Triggering the Conflict
Let's merge `feature-discount` into `main`. Since both branches appended differing functions to the same file positions, Git will halt and raise a conflict.

```python
# Merge branch
!git merge feature-discount
```

```python
# Let's inspect the conflicted file to see Git's markers
with open("app.py", "r") as f:
    print(f.read())
```

### 3.2: Resolving the Conflict
To resolve the conflict, we must open `app.py`, remove the conflict markers (`<<<<<<<`, `=======`, `>>>>>>>`), and choose the final calculation code. 

Let's write a clean, resolved version of `app.py` keeping the 10% percentage discount, and clean up the syntax.

```python
# Rewrite resolved file contents
resolved_code = """print("=== COIMBATORE CANTEEN BILLING v1.0.0 ===")
price = 15
count = 3
print(f"Total: {price * count}")

def apply_discount(total):
    return total * 0.90  # Resolved: keeping 10% percentage discount

print(f"Final with discount: {apply_discount(price * count)}")
"""

with open("app.py", "w") as f:
    f.write(resolved_code)

print("Conflict resolved in app.py")
```

```python
# Stage and commit the resolution
!git add app.py
!git commit -m "merge: resolve discount conflict by choosing percentage rate"

# Verify the repository is clean
!git status
```

```python
# Check the commit history graph
!git log --oneline --graph
```
